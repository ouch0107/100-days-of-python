```python
from ipywidgets import interact
```

```python
def plot(a):
    b, g, r = cv2.split(img)
    alpha = np.ones(b.shape, dtype=b.dtype) * a
    img_RGBA = cv2.merge((r, g, b, alpha))
    plt.imshow(img_RGBA)
```

```python
interact(plot, a=(0, 255, 5))
```
